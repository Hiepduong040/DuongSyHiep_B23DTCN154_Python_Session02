```python
# Nhập số nguyên dương n từ người dùng
n = int(input("Nhập một số nguyên dương n: "))

# Kiểm tra xem n có phải là số nguyên dương không
if n < 0:
    print("Vui lòng nhập một số nguyên dương.")
else:
    # Tính giai thừa n!
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    
    # In ra kết quả
    print(f"{n}! = {factorial}")
```

```python
# bai 5:
n = int(input("Nhập số nguyên dương n: "))

if n > 0:
    so_goc = n
    so_dao = 0
    
    while n > 0:
        chu_so = n % 10             # Lấy chữ số cuối
        so_dao = so_dao * 10 + chu_so  # Thêm vào số đảo
        n = n // 10                 # Bỏ chữ số cuối
    
    if so_goc == so_dao:
        print(f"{so_goc} là số đối xứng.")
    else:
        print(f"{so_goc} không phải số đối xứng.")
else:
    print("Vui lòng nhập số nguyên dương!")
```

```python
#Bài 7:
def is_prime(num):
    """Kiểm tra xem số num có phải là số nguyên tố không."""
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Nhập hai số nguyên a và b từ người dùng
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))

# Đảm bảo a là nhỏ hơn hoặc bằng b
if a > b:
    a, b = b, a  

print(f"Các số nguyên tố trong khoảng [{a}, {b}] là:")
for number in range(a, b + 1):
    if is_prime(number):
        print(number, end=' ')
```
